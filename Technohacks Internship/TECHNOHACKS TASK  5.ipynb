### Removing Duplicates

```python
df.drop_duplicates(subset = "Id" , keep = "first")
df.drop_duplicates(subset = "SepalLengthCm" , keep = "first")
df.drop_duplicates(subset = "SepalWidthCm" , keep = "first")
df.drop_duplicates(subset = "PetalLengthCm" , keep = "first")

```
